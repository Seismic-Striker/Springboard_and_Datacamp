```python
#import modules that will be helpful in our exploration
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

#import dataset

# Assign the filename: file
file = 'chesswithranks.csv'
# Read the file into a DataFrame: df
df = pd.read_csv(file)
df.head(2)
```

```python
bronze = df[df.white_rank == 'Bronze'] 
silver = df[df.white_rank == 'Silver'] 
gold = df[df.white_rank == 'Gold'] 
plat = df[df.white_rank == 'Bronze'] 
diamond = df[df.white_rank == 'Diamond'] 
onyx = df[df.white_rank == 'Onyx'] 
```

```python
bronze.groupby('opening_shortname')['game_id'].count().sort_values(ascending = False).head(15).plot(kind = 'barh')
```

```python
categories = [bronze,silver,gold,plat,diamond,onyx]
for cat in categories:
    cat.groupby('opening_shortname')['game_id'].count().sort_values(ascending = False).head(15).plot(kind = 'barh')
    print(cat.white_rank.value_counts())
    plt.show()
```
