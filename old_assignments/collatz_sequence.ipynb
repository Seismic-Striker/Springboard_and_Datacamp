```python
import numpy as np

import matplotlib.pyplot as plt

def collatz_sequence(x):
    num_seq = [x]
    if x < 1:
       return []
    while x > 1:
       if x % 2 == 0:
         x = x / 2
       else:
         x = 3 * x + 1
       num_seq.append(x)    
    return num_seq

x = [x for x in range(len(collatz_sequence(9663)))]

y = (collatz_sequence(9663))

x= np.array(x)

y = np.array(y)

print(f'Max index is : {np.argmax(y)}')

print(y)

plt.plot(x,y)


```
